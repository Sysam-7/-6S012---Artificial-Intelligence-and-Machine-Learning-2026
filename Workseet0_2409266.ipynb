```python
#4. Remove the First and Last Elements
def remove_first_last(lst):
    """
    Removes the first and last element from a list.

    Parameters:
        lst (list): Input list

    Returns:
        list: List without first and last elements
    """
    return lst[1:-1]


# Example
print(remove_first_last([1, 2, 3, 4, 5]))
```

```python
#5. Get the First n Elements
def get_first_n(lst, n):
    """
    Returns first n elements of a list.
    """
    return lst[:n]


# Example
print(get_first_n([1, 2, 3, 4, 5], 3))
```

```python
#6. Extract Last n Elements
def get_last_n(lst, n):
    """
    Returns last n elements of a list.
    """
    return lst[-n:]


# Example
print(get_last_n([1, 2, 3, 4, 5], 2))
```

```python
#7. Reverse Skip (Every Second Element Backwards)
def reverse_skip(lst):
    """
    Returns every second element from the second-to-last in reverse.
    """
    return lst[-2::-2]


# Example
print(reverse_skip([1, 2, 3, 4, 5, 6]))
```

4.3 Nested List Exercises

```python
#1. Flatten a Nested List
def flatten(lst):
    """
    Flattens a nested list by one level.
    """
    result = []

    for sublist in lst:
        for item in sublist:
            result.append(item)

    return result


# Example
print(flatten([[1,2],[3,4],[5]]))
```

```python
#2. Access Nested Element
def access_nested_element(lst, indices):
    """
    Access element in nested list using indices.
    """
    element = lst
    for i in indices:
        element = element[i]
    return element


# Example
lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst, [1,2]))
```

```python
#3. Sum of All Elements in Nested List
def sum_nested(lst):
    """
    Recursively sums elements of nested lists.
    """
    total = 0

    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item

    return total


# Example
print(sum_nested([[1,2],[3,[4,5]],6]))
```

```python
#4. Remove Specific Element
def remove_element(lst, elem):
    """
    Removes all occurrences of elem from nested list.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)

    return result


# Example
print(remove_element([[1,2],[3,2],[4,5]], 2))
```

```python
#5. Find Maximum in Nested List
def find_max(lst):
    """
    Finds maximum value in nested list.
    """
    max_val = float('-inf')

    for item in lst:
        if isinstance(item, list):
            max_val = max(max_val, find_max(item))
        else:
            max_val = max(max_val, item)

    return max_val


# Example
print(find_max([[1,2],[3,[4,5]],6]))
```

```python
#6. Count Occurrences in Nested List
def count_occurrences(lst, elem):
    """
    Counts occurrences of element in nested list.
    """
    count = 0

    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1

    return count


# Example
print(count_occurrences([[1,2],[2,3],[2,4]], 2))
```

```python
#7. Deep Flatten
def deep_flatten(lst):
    """
    Flattens list of any depth.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)

    return result


# Example
print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

```python
#8. Average of Nested List
def average_nested(lst):
    """
    Returns average of nested list values.
    """
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)


# Example
print(average_nested([[1,2],[3,4],[5,6]]))
```
